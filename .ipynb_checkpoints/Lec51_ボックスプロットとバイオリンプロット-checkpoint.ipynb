```python
import numpy as np
from numpy.random import randn
from scipy import stats 
import seaborn as sns
    
%matplotlib inline
```

```python
data1 =randn(100)
data2 =randn(100) + 2
```

```python
sns.distplot(data1)
sns.distplot(data2)
```

```python
sns.boxplot(data=[data1,data2])

#箱ひげ図
#真ん中が中央値,箱は２５〜７５％、
```

```python
sns.boxplot(data=[data1,data2],whis=np.inf)

#外れ値も含んでひげも伸びる
```

```python
sns.boxplot(data=[data1,data2],orient='v')

#horizon=水平、orient=方向
```

```python
data1=stats.norm(0,5).rvs(100)
```

```python
data2=np.concatenate([stats.gamma(5).rvs(50)-1,stats.gamma(5).rvs(50)])
```

```python
sns.boxplot(data=[data1,data2])
```

```python
sns.violinplot(data=[data1,data2])
```

```python
sns.violinplot(data=data2,bw=0.01)
```

```python
sns.violinplot(data=data1,inner='stick')

#stick=データがどこにあるのかわかる
```
