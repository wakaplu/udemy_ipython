<b> 1) タイタニック号の乗客はどのような人達だったのか？</b>

```python
#numpyなど図表描画に必要なモジュールをインストール

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#まずは性別

sns.countplot('Sex',data=titanic_df)
```

全体としては男性の方が多い

```python
#乗客の性別を客室で層別化する、Pclass=客室

sns.countplot('Sex', data=titanic_df, hue='Pclass')
```

三等級の客室には男性が多い

```python
#今度は逆に乗客の客室を性別で層別化

sns.countplot('Pclass', data=titanic_df, hue='Sex')
```

やはり三等客室は女性に比べて男性が圧倒的に多い。<br>
今度は子どもについても詳しく見ていく。

```python
#16歳未満を子どもとして定義し直す

def male_female_child(passenger):
    age, sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex
    
titanic_df['person'] =titanic_df[['Age','Sex']].apply(male_female_child, axis=1)
```

```python
#もう一度、客室別に人数を図示

sns.countplot('Pclass',data=titanic_df,hue='person')
```

三等客室にはたくさんの子どもがいるのに対し、  <br>
一等客室にはほとんど子どもがいない。

```python
#年齢のヒストグラムを表示

titanic_df['Age'].hist(bins=70)
```

乳幼児から大人まで幅広くいる。

```python
#平均年齢

titanic_df['Age'].mean()
```

```python
#性別、子どもの人数の小計

titanic_df['person'].value_counts()
```

```python
#性別別のカーネル密度推定のグラフを1つのプロットとしてなめらかに図示

fig = sns.FacetGrid(titanic_df, hue='Sex', aspect=4)
fig.map(sns.kdeplot,'Age',shade=True)
oldest = titanic_df['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
#子ども（赤）についても追加

fig = sns.FacetGrid(titanic_df, hue='person', aspect=4)
fig.map(sns.kdeplot,'Age',shade=True)
oldest = titanic_df['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
#客層別だとこんな感じ

fig = sns.FacetGrid(titanic_df, hue='Pclass', aspect=4)
fig.map(sns.kdeplot,'Age',shade=True)
oldest = titanic_df['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

若い層のほうが2、３等級客室多い<br>
１等級は幅広い年齢だが、やはり年齢層高めの人が多いか

<b>2)それぞれの乗客はどのデッキにいたか？また、それは客室の種類とどのような関係にあったか？</b>

```python
#デッキの場所に関する欠損地を削除しdeckという新しいオブジェクトを作成

deck =titanic_df['Cabin'].dropna()
```

```python
#どのデッキにいたかに関する情報を表示
#最初の１文字（アルファベットさえわかればどこにいたのかわかりそう）

deck.head()
```

```python
#for文で１文字目を取り出す

levels =[]
for level in deck:
    levels.append(level[0])
```

```python
#DataFrameを使ってプロット

cabin_df =DataFrame(levels)
cabin_df.columns =['Cabin']
```

```python
cabin_df.head()
```

アルファベットだけ取り出せた

```python
#乗客がどこにいたのか図示

sns.countplot('Cabin',data=cabin_df,palette='winter_d',order=sorted(set(levels)))
```

ん？Tってなんだこれ<br>
おそらく間違っているデータなのでこれを外して新しいデータを作ります

```python
#Tを外す

cabin_df =cabin_df[cabin_df.Cabin != 'T']
```

```python
#改めて

sns.countplot('Cabin',data=cabin_df,palette='summer',order=sorted(set(cabin_df.Cabin)))
```

乗客はC地点にいた人が最も多く、おそらく船の真ん中あたりに密集していた？

<b>4)家族づれか？単身者か？</b>

```python
# 単身者かどうかを示す新しい列を追加

# 親子、兄弟姉妹を示す列を足し合わせる
titanic_df['Alone'] =  titanic_df.Parch + titanic_df.SibSp
titanic_df['Alone'].head()
```

```python
#数字を文字列に変換

titanic_df['Alone'].loc[titanic_df['Alone']>0] = 'With Family'
titanic_df['Alone'].loc[titanic_df['Alone']==0] = 'Alone'
```

```python
#確認

titanic_df.head()
```

```python
#単身者か家族づれか図示

sns.countplot('Alone',data=titanic_df,palette='Blues')
```

独り身の方が多い

<b>5) 沈没からの生還者には、どのような要因があったのか？</b>

```python
#「Suvivor」という新しい列を作る

titanic_df['Survivor']=titanic_df.Survived.map({0:'no',1:'yes'})
```

```python
#生存したかどうかの図示

sns.countplot('Survivor',data=titanic_df,palette='Set1')
```

犠牲者のほうが多い。<br>
映画タイタニックでは３等客室の乗客への扱いが不当だったが、やはり関係あるのか？

```python
#客室のランクと生存率の図示
sns.factorplot('Pclass','Survived',data=titanic_df,order=[1,2,3])
```

３等客室の生存率はかなり低め。<br>
ただし、先ほど３等客室には男性が多かったことがわかっていたので、<br>
客室というよりは男性が多く犠牲になった可能性もある。

```python
#先ほどのグラフに性別、子どもかどうかも考慮

sns.factorplot('Pclass','Survived',hue='person', data=titanic_df,order=[1,2,3],aspect=2)
```

・男性だと客室のランクに関わらず男性の生存率は低い<br>
・女性と子供の生存率は１、２等客室でそんなに変わらない<br>
・２等客室では「女性と子ども優先」がかなり顕著か？<br>
・第３客室はやはり生存率は全体的に低い

```python
#年齢と性別のグラフを図示

sns.lmplot('Age','Survived',data=titanic_df)
```

年齢が上がるにつれて生存率が下がっている<br>
客室の情報を加えるとどうなるだろうか

```python
#客室の種類で層別化する
sns.lmplot('Age','Survived',hue='Pclass',data=titanic_df,palette='winter',hue_order=[1,2,3])
```

客室が高いほど生存率は高いが、年齢が高いほど生存率は低下

```python
#10代、２０代、40代、６０代、８０代に印をつけて標準偏差を追記

generations = [10,20,40,60,80]
sns.lmplot('Age','Survived',hue='Pclass',data=titanic_df,palette='winter',
          hue_order=[1,2,3],x_bins=generations)
```

1等客室に真っ先に助けられた人がいるのかも.<br>
歳をとった偉い人？

```python
#性別と年齢の関係もみてみる

sns.lmplot('Age','Survived',hue='Sex',data=titanic_df,palette='winter',
          x_bins=generations)
```

・女性は年齢が上がるほど生存率が上がる<br>
・男性は年齢が上がるほど生存率が下がるが、高年齢で飛び抜けて生存率が高い人がいる
