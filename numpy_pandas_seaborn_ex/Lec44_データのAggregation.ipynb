```python
dframe_wine =pd.read_csv('winequality-red.csv',sep=';')
```

```python
dframe_wine.head()
```

```python
dframe_wine['alcohol'].mean()
```

```python
def max_to_min(arr):
    return arr.max() - arr.min()
```

```python
wino =dframe_wine.groupby('quality')
```

```python
wino.describe()
```

```python
wino.agg(max_to_min)
```

```python
wino.agg('mean')
```
